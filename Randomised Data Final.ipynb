```python
import pandas as pd
import random
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
plt.style.use('bmh')
```

```python
num=30
s = pd.DataFrame(np.random.randint(1,100,size=(num, 3)), columns=list('PCM'))
```

```python
s
```

```python
total=s.sum(axis=1)
s['T']=total
s
```

```python
st=s.sort_values(by='T')
st
```

```python
a=s.P/s.P.median(0)
b=s.C/s.C.median(0)
c=s.M/s.M.median(0)
```

```python
plt.bar(list(s.index.values),a)
plt.plot(a,label='Physics')
plt.legend()
plt.show()
```

```python
plt.plot(b,label='Chemistry',color='C1')
plt.bar(list(s.index.values),b)
plt.legend()
plt.show()
```

```python
plt.plot(c,label='Maths',color='C2')
plt.bar(list(s.index.values),c)
plt.legend()
plt.show()
```

```python
s=s.drop('T',axis=1)
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
plt.hist(s,bins=5)
plt.gca().set(title='Marks Frequency Histogram', ylabel='Frequency');
plt.legend(['P','C','M'])
plt.show()
```

```python
print(s.corr())
plt.plot(s.corr())
plt.legend(['P','C','M'])
plt.title("Correlation in marks")
plt.show()
```

```python
print(s.cov())
plt.plot(s.cov())
plt.legend(['P','C','M'])
plt.title("Covarriance in marks")
plt.show()
```

```python
ns=s
```

```python
mode=3*(ns.median())-2*(ns.mean())
mode
```

```python
plt.plot(ns.mean(),label='Mean')
plt.plot(ns.median(),label='Median')
plt.plot(mode,label='Mode')
plt.title("Mean ,Median and Mode of the Data")
plt.legend()
```

```python
print("The Highest marks in each subject are")
s[['P','C','M']].max(0)
```

```python
print("The Lowest marks in each subject are")
s[['P','C','M']].min(0)
```

```python
plt.title("The Highest marks for each student in any subject are")
plt.plot(s[['P','C','M']].max(1))
```

```python
plt.title("The Lowest marks for each student in a subject are")
plt.plot(s[['P','C','M']].min(1))
```

```python
plt.title("Quantile for each subject")
print(ns.quantile())
plt.plot(ns.quantile())
```
