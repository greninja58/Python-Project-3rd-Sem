```python
import pandas as pd
import random
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
plt.style.use('bmh')
```

```python
s = pd.DataFrame(np.random.randint(1,100,size=(30, 3)), columns=list('PCM'))
```

```python
s
```

```python
a=s.P/s.P.median(0)
b=s.C/s.C.median(0)
c=s.M/s.M.median(0)
```

```python
#plt.bar(s,1)
plt.plot(a,label='Physics')
plt.legend()
plt.show()
```

```python
plt.plot(b,label='Chemistry',color='C1')
plt.legend()
plt.show()
```

```python
plt.plot(c,label='Maths',color='C2')

plt.legend()
plt.show()
```

```python
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
plt.hist(s,bins=5)
plt.gca().set(title='Marks Frequency Histogram', ylabel='Frequency');
plt.legend(['P','C','M'])
plt.show()
```

```python
print(s.corr())
plt.plot(s.corr())
plt.legend(['P','C','M'])
plt.show()
```

```python
print(s.cov())
plt.plot(s.cov())
plt.legend(['P','C','M'])
plt.show()
```

```python
ns=s
ns
```

```python
mode=3*(ns.median())-2*(ns.mean())
mode
```

```python
plt.plot(ns.mean(),label='Mean')
plt.plot(ns.median(),label='Median')
plt.plot(mode,label='Mode')
plt.legend()
```
